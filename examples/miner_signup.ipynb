# How to Join Nuance

## 1. Create Wallet
Before you can join Nuance, you need to create a Bittensor coldkey and a hotkey, information on this you can find on: https://docs.bittensor.com/getting-started/wallets


## 2. Register
Then you need to register your miner on sn23, which costs a small registration fee paid in Tao. More info you find at: https://docs.bittensor.com/miners#miner-registration

## 3. Verify Account
Afterwards, you need to verify your X account.

Ensure your X account is public and has posting activity.

Create a new post that quotes the Nuance announcement post: https://x.com/NuanceSubnet/status/1909263356654952674

Include your hotkey address in the post text.

Example:
```
I'm joining the Nuance Network as a miner with hotkey: 5F7nTtN...XhVVi
```
Save the verification post ID as you'll need it for the next step.

## 4. Run the Notebook to Push Account to Chain

In order for valdiators to detect your account, you need to push the post id on chain.

The code below installs Bittensor, let's you regenerate your wallet using only your public coldkey and your hotkey, and then let's you push your verification post to chain.

````python
import bittensor as bt

x_account_username = "your_account_username"
verification_post_id = "the_id_of_your_post"
WALLET_NAME = "default"
HOTKEY = "default"

subtensor = bt.Subtensor(network="finney")

wallet = bt.wallet(name=WALLET_NAME, hotkey = HOTKEY)

commit_data = f"{x_account_username}@{verification_post_id}"
subtensor.commit(wallet=wallet, netuid=23, data=commit_data)
````
